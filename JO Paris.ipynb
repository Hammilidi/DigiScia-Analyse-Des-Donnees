Ces statistiques descriptives nous permettent de tirer plusieurs interprtations et conclusions concernant la distribution des mdailles parmi les pays du dataset des Jeux Olympiques de Paris.

### Interprtations :

   - **Gold (1.00)** : La mdiane montre que la moiti des pays ont remport au moins 1 mdaille d'or. Cela signifie que beaucoup de pays remportent peu de mdailles d'or.

### Conclusions :

3. **Bronze Plus Rpandu** :

La matrice de corrlation fournie montre les coefficients de corrlation entre les mdailles d'or, d'argent et de bronze. Chaque coefficient de corrlation mesure la force et la direction de la relation linaire entre deux variables. Ces valeurs sont comprises entre -1 et 1 :

- **1** signifie une corrlation parfaitement positive (les variables augmentent ensemble).
- **0** signifie aucune corrlation (aucune relation linaire).
- **-1** signifie une corrlation parfaitement ngative (l'une augmente tandis que l'autre diminue).
