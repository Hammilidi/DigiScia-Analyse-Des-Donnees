### Qu'est-ce que Pandas ?

Pandas est une bibliothèque Python utilisée pour travailler avec des ensembles de données.

Il dispose de fonctions d'analyse, de nettoyage, d'exploration et de manipulation des données.

Le nom « Pandas » fait référence à la fois à « Panel Data » et à « Python Data Analysis » et a été créé par Wes McKinney en 2008.

### Pourquoi utiliser Pandas ?

Pandas nous permet d’analyser le big data et de tirer des conclusions basées sur des théories statistiques.

Les pandas peuvent nettoyer des ensembles de données en désordre et les rendre lisibles et pertinents.

Les données pertinentes sont très importantes en science des données.

# Nettoyage

```python
- Apres une analyse exploration, il est crucial de proceder au nettoyage de vos donnees. 
- Cette etape consiste a epurer vos donnees, en corrigeant les erreurs, les incoherences, et en ne conservan
```

```python
# Gestion des valeurs manquantes
# df.isna().sum()
# df.dropna(axis=0, inplace=True)  # Supprime les lignes avec des valeurs manquantes
# df.dropna(axis=1, inplace=True)  # Supprime les colonnes avec des valeurs manquantes

# df.fillna(df.mean(), inplace=True)  # Remplit avec la moyenne des colonnes
# df.fillna(method='ffill', inplace=True)  # Remplit avec la dernière valeur connue

df.isnull().sum()
df.isna().sum()

df.dropna(axis=0)
df.dropna(axis=1)
df = df.fillna(df.mean())
```

```python
# Duplications 
df.duplicated().sum()

df.drop_duplicates(inplace=True)
```

```python
# Renommer des colonnes
# df.rename(columns={'old_name': 'new_name'}, inplace=True)
df.rename(columns={'Sex': 'Gender'}, inplace=True)
df.columns
```

```python
# Filter des donnees 
# df = df[df['Age'] >= 18]  # Supprime les lignes où 'column_name' est <= 0
# df[df.isna()].sum()
filtre = df["Age"]>=18
df[filtre]
```

```python
# Convertir le type d'une colonne
df['Fare'] = df['Fare'].astype(int)
df.dtypes
```

# Analyses avec Pandas

```python
print(df['Age'].count()) # nombre de lignes de la colonne Age
print(df['Age'].min()) # L'age minimal
print(df['Age'].max()) # L'age maximal
print(df['Age'].mean()) # L'age moyen
print(df['Age'].median()) # L'age median
print(df['Age'].sum()) # La somme de tous les ages 
print(df['Age'].mode()) # Le mode c'est-a-dire l'age qui se repete le plus 

print("\n\n")

# Avoir plusieurs statistiques a la fois avec la methode .agg() pour aggregation pour une seule colonne
print(df['Age'].agg(['mean', 'median', 'min', 'max', 'sum', 'std']))

print("\n\n")

# Avoir plusieurs statistiques a la fois avec la methode .agg() pour aggregation pour une plusieurs colonnes
df.agg({'Age':['mean', 'median', 'min', 'max', 'sum'], 'Fare': ['mean', 'median', 'min', 'max', 'sum', 'std']})
```

### Pie Chart

```python
# Créer un diagramme circulaire
plt.figure(figsize=(4, 4))
explode = (.1, 0) # exploser la permiere portion du diagramme
# len(df.groupby('Sex')['Fare'].mean())
plt.pie(df.groupby('Sex')['Fare'].mean(), labels=df['Sex'].unique(), autopct='%.2f %%', explode= explode) # la methode .pie() permet d'affciher le diagramme circulaire. Le permier argument correspond aux prix moyens par genre. labels pour affciher les libelles de chaque portion du diagramme. autopct pour precisr le nombre de valeurs a prendre apres la virgule  
# plt.legend()
plt.title("Prix Moyen Par Genre")

plt.show()
```

### Diagramme a moustaches 

- Le diagramme a moustaches ou box plot permet d'afficher les quantiles (25%, 50% ou la mediane, 75%).
